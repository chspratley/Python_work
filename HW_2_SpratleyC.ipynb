```python
# Added in Numpy library
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Dataset: Feature (sq. ft) and Target (price)
X = np.array([1500, 1600, 1700, 1800, 1900]).reshape(-1,1)
y = np.array([300000, 320000, 340000, 360000, 380000])

# Initiate intercept
X_b = np.c_[np.ones((X.shape[0], 1)), X]

# Setup cost function and add meaningful names to variables
def compute_cost(X, y, theta):
    num_items = len(y)
    hypothesis = X.dot(theta)
    cost_function = 1 / (2 * num_items) * np.sum((hypothesis - y) ** 2)
    return cost_function

# Add on two additional sets,theta 0 and theta 1
theta_values = [ np.array([100000, 150]),
                 np.array([120000, 130]),
                 np.array([110000, 140]),
                 np.array([130000, 120]),
                 np.array([140000, 110])]

# Plot the data
plt.figure(figsize=(10, 6))
plt.scatter(X, y,color="blue", label="Training Data")

# Create all new values for the variable passed in the function
for i, theta in enumerate(theta_values):

  # Compute the cost of the parameters
  cost = compute_cost(X_b, y, theta)

  # Print only the cost
  print(f"The cost for theta {i + 1} is: {cost}")

  # Compute predictions
  hx = X_b.dot(theta)
  plt.plot(X, hx, label=f"Fitted Line {i + 1}: θ0 = {theta[0]}, θ1 = {theta[1]}")

# Add Desription to label and title to reflect the values and purpose
plt.xlabel("Feature (sq. ft)")
plt.ylabel("Target (price)")
plt.title ("Effects of the Cost function using different Parameters")
plt.legend()
plt.grid(True)
plt.show()
```
